```python
def cv_texto(c):
    secciones = [
        ("Curriculum Numero: ", c["id"]),
        ("Nivel y perfil", f'{c["nivel"]} | {c["perfil"]} | {c["anios"]} años de experiencia'),
        ("Resumen", c["resumen"]),
        ("Formación académica", c["formacion"]),
        ("Experiencia laboral", c["experiencia"]),
        ("Habilidades técnicas", ", ".join(c["tecnicas"])),
        ("Habilidades blandas", ", ".join(c["blandas"])),
        ("Certificaciones", ", ".join(c["certificaciones"])),
        ("Idiomas", ", ".join(c["idiomas"])),
        ("Proyectos relevantes", "; ".join(c["proyectos"])),
    ]
    return "\n".join(f"{k}: {v}" for k,v in secciones)

def curriculums_pdf(path, text):
    c = canvas.Canvas(str(path), pagesize=letter)
    width, height = letter
    y = height - 48

    for line in text.splitlines():
        c.drawString(48, y, line)
        y -= 12
    c.save()

def curriculums_png(path, text):
    img = Image.new("RGB", (1400, 1800), "white")
    d = ImageDraw.Draw(img)

    y = 60
    for raw in text.splitlines():
        for line in textwrap.wrap(raw, width=92):
            d.text((70, y), line, fill="black")
            y += 32

    meta = PngImagePlugin.PngInfo()
    meta.add_text("cv_text", text)

    img.save(path, pnginfo=meta)

png_ids = {"C04", "C08", "C13", "C16"}  # Para crear 1 Junior, 2 Intermedio, 1 Senior
manifest = []
for _, c in df_candidatos.iterrows():
    ext = ".png" if c["id"] in png_ids else ".pdf" # Para identificar si debe ser pdf o png
    path = CV_DIR / f'{c["id"]}{ext}'
    text = cv_texto(c)
    if ext == ".png":
        curriculums_png(path, text)
    else:
        curriculums_pdf(path, text)
    manifest.append({
        "id":c["id"],
        "nivel":c["nivel"],
        "perfil":c["perfil"],
        "archivo":path.name,
        "formato":ext[1:].upper()
    })

mf = pd.DataFrame(manifest)
print(pd.crosstab(
    mf.nivel,
    mf.formato)
.to_string())
print("\nTotal archivos", len(list(CV_DIR.iterdir()))) # Debe crear los 20 cvs
```

```python
def extraer_info(path):
    if path.suffix.lower() == ".pdf":
        return "\n".join(
            (p.extract_text() or "")
            for p in PdfReader(str(path)).pages
        )

    if path.suffix.lower() == ".png":
        with Image.open(path) as im:
            return im.info.get("cv_text", "")

documentos = []

for row in manifest:
    # print("Si entre al for")
    texto = extraer_info(CV_DIR / row["archivo"])
    documentos.append({
        **row,
        "texto": texto
    })

assert len(documentos) == 20
print("Documentos procesados:", len(documentos))
```

```python
def normalizar_texto(texto: str) -> str:
    if not texto:
        return ""
    texto = texto.lower()
    texto = unicodedata.normalize("NFKD", texto)
    texto = "".join(ch for ch in texto if not unicodedata.combining(ch))
    texto = re.sub(r"\s+", " ", texto).strip()
    return texto

for doc in documentos:
    doc["texto_norm"] = normalizar_texto(doc["texto"])

#print(documentos[0]["texto_norm"][:300])
print(documentos[11]["texto_norm"][:300])
```

```python
!pip install faiss-cpu
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize
import faiss
import numpy as np
```

```python
corpus_cv = [info["texto_normalizado"] for info in info_estructurada.values()]
corpus_completo = corpus_cv + [vacante_texto_norm]

tfidf_vectorizer = TfidfVectorizer(
    max_features=2000,
    ngram_range=(1, 2),
    min_df=1,
)
tfidf_vectorizer.fit(corpus_completo)

#print('bug')
ids_cv = list(info_estructurada.keys())
matriz_cv = tfidf_vectorizer.transform(
    [info_estructurada[i]["texto_normalizado"] for i in ids_cv]
).toarray().astype("float32")
matriz_cv = normalize(matriz_cv, norm="l2", axis=1)

dimension = matriz_cv.shape[1]
indice_faiss = faiss.IndexFlatIP(dimension)
indice_faiss.add(matriz_cv)

metadatos_por_posicion = [
    {
        "id": i,
        "nivel": info_estructurada[i]["nivel"],
        "perfil": info_estructurada[i]["perfil"],
        "anios_experiencia": info_estructurada[i]["anios_experiencia"],
        "formato_archivo": info_estructurada[i]["formato_archivo"],
    }
    for i in ids_cv
]

print(f"Base de datos vectorial - incluye {indice_faiss.ntotal} Curriculums.")
print(f"Dimensión de cada vector {dimension}")
```

```python
# MUSTTT - Vectoriza el texto del puesto
vector_vacante = tfidf_vectorizer.transform([vacante_texto_norm]).toarray().astype("float32")
vector_vacante = normalize(vector_vacante, norm="l2", axis=1)
```

```python
K = 10  # utilizamos mas para poder rankearlos despues
similitudes, posiciones = indice_faiss.search(vector_vacante, K)

similitudes = similitudes[0]
posiciones = posiciones[0]

df_busqueda = pd.DataFrame({
    "id curriculum": [metadatos_por_posicion[p]["id"] for p in posiciones],
    "match": similitudes,
    "seniority_nivel": [metadatos_por_posicion[p]["nivel"] for p in posiciones],
    "perfil": [metadatos_por_posicion[p]["perfil"] for p in posiciones],
    "anios_experiencia": [metadatos_por_posicion[p]["anios_experiencia"] for p in posiciones],
    "formato_archivo": [metadatos_por_posicion[p]["formato_archivo"] for p in posiciones],
})

df_busqueda.sort_values("match", ascending=False)
```

```python
KEYWORDS_VACANTE = {
    "python", "c++", "c#", "java", "javascript",
    "machine learning", "nlp", "llm", "apis", "api",
    "data pipelines", "pipelines", "azure", "cloud",
    "distributed systems", "microservices", "docker", "kubernetes",
    "monitoring", "ci/cd", "vector databases", "sql",
}

def normalizar_lista(lista):
    return {normalizar_texto(x) for x in lista}

def calcular_match_keywords(info_candidato):
    habilidades = normalizar_lista(info_candidato["habilidades_tecnicas"])
    certificaciones = normalizar_lista(info_candidato["certificaciones"])
    bolsa_candidato = habilidades | certificaciones

    coincidencias = set()
    for kw in KEYWORDS_VACANTE:
        for term in bolsa_candidato:
            if kw in term or term in kw:
                coincidencias.add(kw)
                break

    score = len(coincidencias) / len(KEYWORDS_VACANTE)
    return score, sorted(coincidencias)

def score_experiencia(anios, nivel):
    return min(anios, 10) / 10

def generar_justificacion(info_candidato, similitud, score_kw, terminos_comunes, score_exp):
    return (
        f"Candidato {info_candidato['id']} ({info_candidato['nivel']}, "
        f"perfil: {info_candidato['perfil']}, {info_candidato['anios_experiencia']} años "
        f"de experiencia). Similitud semántica con la vacante: {similitud:.2f}. "
        f"Coincidencia de habilidades técnicas/certificaciones relevantes "
        f"({len(terminos_comunes)} de {len(KEYWORDS_VACANTE)} palabras clave): "
        f"{', '.join(terminos_comunes) if terminos_comunes else 'ninguna coincidencia directa'}. "
        f"Formación: {info_candidato['formacion'].rstrip('.')}. "
        f"Proyecto destacado: {info_candidato['proyectos'][0] if info_candidato['proyectos'] else 'N/D'}."
    )

filas_ranking = []
for cid, sim in zip(df_busqueda["id curriculum"], df_busqueda["match"]):
    info_c = info_estructurada[cid]
    score_kw, terminos = calcular_match_keywords(info_c)
    score_exp = score_experiencia(info_c["anios_experiencia"], info_c["nivel"])

    # Puntaje final ponderado: 50% similitud semántica, 35% match de habilidades/certificaciones, 15% experiencia
    score_final = 0.50 * sim + 0.35 * score_kw + 0.15 * score_exp

    filas_ranking.append({
        "id": cid,
        "nivel": info_c["nivel"],
        "perfil": info_c["perfil"],
        "anios_experiencia": info_c["anios_experiencia"],
        "similitud_semantica": round(sim, 3),
        "score_keywords": round(score_kw, 3),
        "score_experiencia": round(score_exp, 3),
        "score_final": round(score_final, 3),
        "justificacion": generar_justificacion(info_c, sim, score_kw, terminos, score_exp),
    })

df_ranking = pd.DataFrame(filas_ranking).sort_values("score_final", ascending=False)
top5 = df_ranking.head(5).reset_index(drop=True)
top5[["id", "nivel", "perfil", "anios_experiencia", "score_final"]]
```

# 8. Reporte tecnico

**Arquitectura**

El sistema usa un flujo de recuperación y re-ranking. Primero se generan los 20 CVs sintéticos en PDF y algunos en PNG, se extrae el texto de cada uno (pypdf para PDF, y metadato embebido para los que son en formato PNG) y se normaliza el texto ( todo el texto en minúsculas, sin acentos).

Para los CV en PNG, en vez de usar OCR real, guardé el texto original
dentro de la metadata de la imagen y lo leo de ahí. Esto simula un OCR perfecto
(sin errores de reconocimiento), lo que hace más sencilla la actividad pero
podría no representar cómo se comportaría el sistema con CVs escaneados
reales que el OCR sí pudiera tener errores.

Con ese texto se entrena un vectorizador TF-IDF (unigramas y bigramas) sobre el corpus de CVs y la vacante, y se construye un índice FAISS con los vectores normalizados, usando producto punto como similitud coseno. Para buscar, se ocupo vectorizar el texto de la vacante de la misma forma y se consulta el índice, se seleccionan los 10 CVs más similares a la vacante. De esos 10 se hace un segundo cálculo que es con base a un score, que basicamente combina la similitud semántica (50%), el match de habilidades técnicas/certificaciones contra palabras clave de la vacante (35%) y los años de experiencia (15%). Con ese score se ordenan y se toman los 5 mejores scores.
Se usó FAISS en lugar de algun LLM por API ya que todo el ranking es con reglas y TF-IDF, para no depender de internet ni de una key.

**Cómo se comportó**

En las pruebas el top 5 salió principalmente por candidatos de perfiles de ML/IA, Datos y Software con experiencia intermedia ya que uno de los datos es que este es un puesto IC2, que es justo lo que se espera para esta vacante. Los perfiles más alejados (BI con Excel, desarrollo móvil) no se metieron en el top 5 lo que nos indica que el score sí está discriminando razonablemente bien.

**Evaluación como experto**

El sistema funciona como un primer filtro decente pero no es perfecto pues puede haber palabras relacionadas pero como mencion en los puestos que tienen los candidatos y no necesariamente como habilidad. Otro principal problema es que TF-IDF solo compara palabras, no entiende sinónimos (por ejemplo "líder de equipo" y "liderazgo" no las relaciona). También los pesos del score (50/35/15) los puse yo a mi criterio personal, sin ninguna validación real de que esos porcentajes sean los correctos con algun reclutador. Con CVs reales que tienen redacción más desordenada probablemente el desempeño bajaría un poco respecto a lo que se vio aquí con CVs sintéticos y que seguian un mismo formato.


**Mejoras propuestas**

* Cambiar TF-IDF por embeddings semánticos (sentence-transformers o un LLM vía API) para que el sistema entienda sinónimos y no solo coincidencia exacta de las palabras.
* Calibrar los pesos del score (actualmente 50/35/15 puestos a mi criterio personal) con datos reales, por ejemplo pidiendo que los establescan un grupo de reclutadores.
* Hacer OCR real sobre los PNG en vez del atajo que usé (leer el texto desde el metadato de la imagen), para que el sistema funcione igual con CVs escaneados reales.
* Agregar más CVs y de diferentes formatos, porque solo 20 es difícil saber qué tan bien generalizaría el sistema con un volumen real de candidatos.
